```python
best_model_juliette = evaluation_df_juliette[evaluation_df_juliette["model_rank"] == 1]
best_model_alexandre = evaluation_df_alexandre[evaluation_df_alexandre["model_rank"] == 1]
best_model_loris = evaluation_df_loris[evaluation_df_loris["model_rank"] == 1]
best_model_paul = evaluation_df_paul[evaluation_df_paul["model_rank"] == 1]

common_models = set(best_model_juliette["model"]) \
                .intersection(best_model_alexandre["model"]) \
                .intersection(best_model_loris["model"]) \
                .intersection(best_model_paul["model"])

common_models_juliette = best_model_juliette[best_model_juliette["model"].isin(common_models)]
common_models_alexandre = best_model_alexandre[best_model_alexandre["model"].isin(common_models)]
common_models_loris = best_model_loris[best_model_loris["model"].isin(common_models)]
common_models_paul = best_model_paul[best_model_paul["model"].isin(common_models)]

concat_all_common_models = pd.concat([common_models_juliette, common_models_alexandre, common_models_loris, common_models_paul], ignore_index=True)
```

```python
best_models = evaluation_df[evaluation_df["model"].isin(concat_all_common_models["model"])]

best_system_prompt = best_models[best_models["system_prompt_rank"] == 1]
best_system_prompts = best_models[best_models["system_prompt_id"].isin(best_system_prompt["system_prompt_id"])]

best_preprompt = best_system_prompts[best_system_prompts["preprompt_rank"] == 1]

best_preprompt = best_preprompt[["model", "system_prompt_id", "preprompt_id",]]
best_preprompt = best_preprompt.drop_duplicates()
best_preprompt.sort_values(by=["preprompt_id", "system_prompt_id", "model"])
```

```python
best_model = evaluation_df[evaluation_df["model_rank"] == 1]
best_models = evaluation_df[evaluation_df["model"].isin(best_model["model"])]

best_system_prompt = best_models[best_models["system_prompt_rank"] == 1]
best_system_prompts = best_models[best_models["system_prompt_id"].isin(best_system_prompt["system_prompt_id"])]

best_preprompt = best_system_prompts[best_system_prompts["preprompt_rank"] == 1]

best_preprompt = best_preprompt[["model", "system_prompt_id", "preprompt_id"]]
best_preprompt = best_preprompt.drop_duplicates()
best_preprompt["model"] = best_preprompt["model"].str.split("/").str[1]
best_preprompt.sort_values(by=["model", "preprompt_id", "system_prompt_id"])
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

colors = ["red", "coral", "gold", "lightgreen","g", "cyan", "paleturquoise", "mediumorchid", "lightpink", "peachpuff", "saddlebrown"]

# Only keep the rows where time is lower than 60s:
fastest_models_df = evaluation_df[evaluation_df["time"] < 60]
for (model_name, group_data), color in zip(fastest_models_df.groupby("model_name"), colors):
    ax.scatter(group_data["time"], group_data["parsed_response_length"], label=model_name, color=color, s=5)

# Set vertical line at 12.5s and 34.2s (average time to open ChatGPT with no prompt and not opened ChatGPT at first with prompt we calculated)
plt.axvline(x=12.5, color="black", linestyle="-.", label="Opened ChatGPT with no prompt: 12.5s, ", linewidth=1)
plt.axvline(x=34.2, color="black", linestyle="--", label="Not opened ChatGPT at first with prompt: 34.2s", linewidth=1)
ax.set_xlabel("Time")
ax.set_ylabel("Parsed Response Length")
ax.legend(title="Model")
plt.title("Time Response per Parsed Response Length per Model")
plt.show()
```
